```
import math

# ==========================================
# GEOMETRY MODULE FUNCTIONS
# ==========================================

def cube_calc(a):
    csa = 4 * (a**2)
    tsa = 6 * (a**2)
    vol = a**3
    return csa, tsa, vol

def cuboid_calc(l, w, h):
    csa = 2 * h * (l + w)
    tsa = 2 * (l*w + w*h + h*l)
    vol = l * w * h
    return csa, tsa, vol

def cylinder_calc(r, h):
    csa = 2 * math.pi * r * h
    tsa = 2 * math.pi * r * (r + h)
    vol = math.pi * (r**2) * h
    return csa, tsa, vol

def cone_calc(r, h):
    # Calculate slant height (l) first
    l = math.sqrt(r**2 + h**2)
    csa = math.pi * r * l
    tsa = math.pi * r * (l + r)
    vol = (1/3) * math.pi * (r**2) * h
    return csa, tsa, vol, l

def sphere_calc(r):
    # Sphere CSA and TSA are the same
    tsa = 4 * math.pi * (r**2)
    vol = (4/3) * math.pi * (r**3)
    return tsa, vol

def hemisphere_calc(r):
    csa = 2 * math.pi * (r**2)
    tsa = 3 * math.pi * (r**2)
    vol = (2/3) * math.pi * (r**3)
    return csa, tsa, vol

# ==========================================
# MAIN INTERACTIVE PROGRAM
# ==========================================

def main():
    print("--- 3D Figure Calculator ---")
    print("1. Cube\n2. Cuboid\n3. Cylinder\n4. Cone\n5. Sphere\n6. Hemisphere\n0. Exit")

    while True:
        choice = input("\nSelect a shape (0-6): ")

        if choice == '1':
            a = float(input("Enter side length: "))
            csa, tsa, vol = cube_calc(a)
            print(f"Results -> CSA: {csa}, TSA: {tsa}, Volume: {vol}")

        elif choice == '2':
            l = float(input("Enter length: "))
            w = float(input("Enter width: "))
            h = float(input("Enter height: "))
            csa, tsa, vol = cuboid_calc(l, w, h)
            print(f"Results -> CSA: {csa}, TSA: {tsa}, Volume: {vol}")

        elif choice == '3':
            r = float(input("Enter radius: "))
            h = float(input("Enter height: "))
            csa, tsa, vol = cylinder_calc(r, h)
            print(f"Results -> CSA: {csa:.2f}, TSA: {tsa:.2f}, Volume: {vol:.2f}")

        elif choice == '4':
            r = float(input("Enter radius: "))
            h = float(input("Enter vertical height: "))
            csa, tsa, vol, l = cone_calc(r, h)
            print(f"Results -> Slant Height: {l:.2f}, CSA: {csa:.2f}, TSA: {tsa:.2f}, Volume: {vol:.2f}")

        elif choice == '5':
            r = float(input("Enter radius: "))
            tsa, vol = sphere_calc(r)
            print(f"Results -> Surface Area: {tsa:.2f}, Volume: {vol:.2f}")

        elif choice == '6':
            r = float(input("Enter radius: "))
            csa, tsa, vol = hemisphere_calc(r)
            print(f"Results -> CSA: {csa:.2f}, TSA: {tsa:.2f}, Volume: {vol:.2f}")

        elif choice == '0':
            print("Exiting...")
            break
        else:
            print("Invalid choice, try again.")

if __name__ == "__main__":
    main()
```
