```
import math

# ==========================================
# CALCULATION MODULE
# ==========================================

def get_cube(op, side):
    if op == 1: return 4 * (side**2)           # CSA
    if op == 2: return 6 * (side**2)           # TSA
    if op == 3: return side**3                 # Volume

def get_cuboid(op, l, w, h):
    if op == 1: return 2 * h * (l + w)         # CSA
    if op == 2: return 2 * (l*w + w*h + h*l)   # TSA
    if op == 3: return l * w * h               # Volume

def get_cylinder(op, r, h):
    if op == 1: return 2 * math.pi * r * h     # CSA
    if op == 2: return 2 * math.pi * r * (r+h) # TSA
    if op == 3: return math.pi * (r**2) * h    # Volume

def get_cone(op, r, h):
    l = math.sqrt(r**2 + h**2)                 # Slant height
    if op == 1: return math.pi * r * l         # CSA
    if op == 2: return math.pi * r * (l + r)   # TSA
    if op == 3: return (1/3) * math.pi * (r**2) * h

def get_sphere(op, r):
    # Sphere CSA and TSA are the same (4 * pi * r^2)
    if op == 1 or op == 2: return 4 * math.pi * (r**2)
    if op == 3: return (4/3) * math.pi * (r**3)

# ==========================================
# MAIN MENU PROGRAM
# ==========================================

def main():
    while True:
        print("\n--- SELECT FIGURE ---")
        print("1. Cube\n2. Cuboid\n3. Cylinder\n4. Cone\n5. Sphere\n0. Exit")

        fig_choice = int(input("Enter choice (0-5): "))
        if fig_choice == 0:
            print("Goodbye!")
            break

        print("\n--- SELECT OPERATION ---")
        print("1. Curved Surface Area\n2. Total Surface Area\n3. Volume")
        op_choice = int(input("Enter choice (1-3): "))

        result = 0
        label = ""

        # Logic for inputs based on figure
        if fig_choice == 1:
            s = float(input("Enter side: "))
            result = get_cube(op_choice, s)

        elif fig_choice == 2:
            l = float(input("Enter length: "))
            w = float(input("Enter width: "))
            h = float(input("Enter height: "))
            result = get_cuboid(op_choice, l, w, h)

        elif fig_choice == 3:
            r = float(input("Enter radius: "))
            h = float(input("Enter height: "))
            result = get_cylinder(op_choice, r, h)

        elif fig_choice == 4:
            r = float(input("Enter radius: "))
            h = float(input("Enter vertical height: "))
            result = get_cone(op_choice, r, h)

        elif fig_choice == 5:
            r = float(input("Enter radius: "))
            result = get_sphere(op_choice, r)

        # Display Result
        ops_names = {1: "Curved Surface Area", 2: "Total Surface Area", 3: "Volume"}
        print("-" * 30)
        print(f"RESULT ({ops_names[op_choice]}): {result:.2f}")
        print("-" * 30)

if __name__ == "__main__":
    main()
```
